# Euro Daily Exchange Rates

The dataset in this project describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. For exmaple: if the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021.

Let's start by reading in the dataset and inspecting the first and the last five rows to understand the structure of the dataset.

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

The dataset consists of 5699 rows and 41 columns. Each row is a single day between `1999-01-04` and `2021-01-08	`. In 40 columns we have exchange rates of the euro to the 40 different currencies around the world. 

We can see that in some columns there are missing values. The most missing values are in columns:
- `[Greek drachma ]`: 5179 missing values,
- `[Slovenian tolar ]`: 3614 missing values,
- `[Cypriot pound ]`: 3353 missing values,
- `[Maltese lira ]`: 3353 missing values,
- `[Slovak koruna ]`: 3091 missing values,
- `[Estonian kroon ]`: 2569 missing values,
- `[Iceland krona ]`: 2407 missing values.

In 3 columns (`[Iceland krona ] `, `[Romanian leu ] `, `[Turkish lira ]`) we have `float` data type. For the other 38 columns, we have `string` data type.

## Data Cleaning

Before we start creating data visualizations, we'll need to do a bit of data cleaning.

Below, we do the following:

- We rename the currency columns to something easier to type (e.g. `[US dollar ]` to `us_dollar`) and `Period\Unit` column to  `time`.
- We change the `time` column to a `datetime` data type.
- We sort the values by `time` in ascending order.
- We reset the index (and drop the initial index).

```python
exchange_rates.columns = exchange_rates.columns.str.replace('[', '').str.replace(']', '')
exchange_rates.columns = exchange_rates.columns.str.strip().str.replace(' ', '_')
exchange_rates.rename(columns={'Period\\Unit:': 'time'}, inplace=True)
exchange_rates.columns = exchange_rates.columns.str.lower()

exchange_rates['time'] = pd.to_datetime(exchange_rates['time'])
exchange_rates.sort_values('time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

exchange_rates
```

Our focus will be on the exchange rate between the euro and the American dollar. We're going to isolate the `time` and the `us_dollar` columns.

```python
euro_to_dollar = exchange_rates[['time', 'us_dollar']].copy()

euro_to_dollar
```

```python
euro_to_dollar['us_dollar'].value_counts()
```

We can seen that in `us_dollar` columns there are 62 entries with `-` character. We're going to drop all the rows  where the `-` character appears. Lastly, we're going to convert the `us_dollar` column to a `float` data type.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['us_dollar'] != '-'].copy()
euro_to_dollar['us_dollar'] = euro_to_dollar['us_dollar'].astype(float)

euro_to_dollar.info()
```

```python
euro_to_dollar
```

## The Exchange Rate Between the Euro and the American Dollar

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.style as style
style.use('default')

plt.plot(euro_to_dollar['time'], euro_to_dollar['us_dollar'])
plt.show()
```

### Rolling Mean

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

We want to show only the long-term trends, so we're going to use the **rolling mean** (also known as the moving average).

```python
plt.figure(figsize=(12, 12))
plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar['time'], euro_to_dollar['us_dollar'])
plt.title('Original values', fontweight='bold')

for i, rolling_window in zip(range(2, 7), [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['time'], euro_to_dollar['us_dollar'].rolling(rolling_window).mean())
    plt.title('Rolling Window: ' + str(rolling_window), fontweight='bold')
```

Our **rolling window** is going to be equal to **30 days**.

```python
euro_to_dollar['rolling_mean_30'] = euro_to_dollar['us_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['time'], euro_to_dollar['rolling_mean_30'])
plt.show()
```

### Euro-Dollar Exchange Rate During the Coronavirus Pandemic

We're going to show how the euro-dollar rate has changed during the coronavirus pandemic. We're going to show the 2020 data and the 2016-2019 data as a baseline.

```python
import datetime as dt

style.use('fivethirtyeight')

# Creating two new dataframes: one for 2016-2020 period, one for 2020
dollar_baseline = euro_to_dollar.copy(
                       )[(euro_to_dollar['time'].dt.year >= 2016) & (euro_to_dollar['time'].dt.year <= 2020)]
dollar_pandemic = euro_to_dollar.copy()[euro_to_dollar['time'].dt.year == 2020]

# Adding the subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))

# Adding the 2016-2020 baseline
ax1.plot(dollar_baseline['time'], dollar_baseline['rolling_mean_30'], color='#af0b1e', linewidth=2.5, alpha=0.1)

# Highlighting 2020 (coronaviruse pandemic)
ax1.plot(dollar_pandemic['time'], dollar_pandemic['rolling_mean_30'], color='#af0b1e', linewidth=2.5)

# Changing tick labels and y-axis
ax1.set_xticklabels(['', '2016', '', '2018', '', '2020'])
ax1.set_yticks([1.05, 1.1, 1.15, 1.2, 1.25])
ax1.tick_params(axis='x', colors='grey')
ax1.tick_params(axis='y', colors='grey')
ax1.set_ylim([1.04, 1.26])

# Title and subtitle
ax1.text(dt.datetime(2015, 3, 1), 1.285, 'Euro-USD rate increased rapidly during coronavirus pandemic', size=14, weight='bold')
ax1.text(dt.datetime(2015, 3, 1), 1.27, 'Euro-USD exchange rate between 2016 and 2021', size=12)

# Adding transparency to the grid
ax1.grid(alpha=0.4)

# Adding plot for 2020
ax2.plot(dollar_pandemic['time'], dollar_pandemic['rolling_mean_30'], color='#af0b1e', linewidth=2.5)

# Changing tick labels and y-axis
ax2.set_xticklabels(['January', 'March', 'May', 'July', 'September', 'November', 'January'], rotation=45)
ax2.set_yticks([1.05, 1.1, 1.15, 1.2, 1.25])
ax2.tick_params(axis='x', colors='grey')
ax2.tick_params(axis='y', colors='grey')
ax2.set_ylim([1.04, 1.26])

# Highlighting the period with highest exchange rate increase
ax2.axvspan(dt.datetime(2020, 5, 18), dt.datetime(2020, 9, 20), ymin=0.045, ymax=0.955, alpha=0.2, color='grey')

# Adding transparency to the grid
ax2.grid(alpha=0.4)

# Signature
ax2.text(dt.datetime(2019, 11, 1), 0.97, 'Adam Sitko' + ' '*109 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d', size=10)

plt.show()
```

```python
franc_pandemic['relative_value'] = franc_pandemic['rolling_mean_30'] / franc_pandemic['rolling_mean_30'].iloc[0]

franc_pandemic
```

```python
yuan_pandemic['relative_value'] = yuan_pandemic['rolling_mean_30'] / yuan_pandemic['rolling_mean_30'].iloc[0]

yuan_pandemic
```
